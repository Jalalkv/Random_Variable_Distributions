```python
# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
```

```python
def log_transfom_clean_(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN 
X_num["clv_log"] = list(map(log_transfom_clean_, X_num["Customer Lifetime Value"]))
```

```python
sns.distplot(X_num["clv_log"])
plt.show()
```
